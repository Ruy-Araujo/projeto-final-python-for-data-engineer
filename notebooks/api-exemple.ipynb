# Exemplo de consulmo da API coinmarketcap

[Documentação](https://coinmarketcap.com/api/documentation/v1/#section/Quick-Start-Guide)

```python
import httpx
import json
```

```python
BASE_URL = 'https://api.coinmarketcap.com/data-api/v3.1'
```

### Preço do Bitcoin Histórico 

Segundo a documentação, conseguimos extrair os preços de todos os dias em um intervalo definido


| parametro | descricao |
|---|---|
|id|Id da moeda|
|timeStart| Timestamp de Inicio em unix|
|timeEnd| Timestamp de Fim  em unix|
|interval| Intervalo de Tempo do Dado|
|convert| Identificador de conversão para moeda |

```python
ROUTE = '/cryptocurrency/historical'

query_string = {
            "id": 1,
            "timeStart": "1690081200", #2023-07-23
            "timeEnd": "1692759600", # 2023-08-23
            "interval": "1d",
            "convert": "BRL",
}

response = httpx.get(BASE_URL+ROUTE, params=query_string)

# json.dumps apenas para formatar a saida
print(json.dumps(response.json(), indent=4))
```
